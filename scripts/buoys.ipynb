```python
import pandas as pd
from seebuoy import NDBC
import matplotlib.pyplot as plt

# specify plot settings such as figure size and font
plt.rcParams["figure.figsize"] = [20,5]
%config InlineBackend.figure_format = 'retina'

ndbc = NDBC(timeframe="historical")
```

## Pull historical station data

```python
available_44034 = ndbc.available_data(station_id="44034")
standard_44034 = ndbc.get_data("44034", dataset='standard')
standard_44034.to_csv("standard_44034.csv")
```

The selected buoy, [44034 at 44.103 N 68.112 W](https://www.ndbc.noaa.gov/station_page.php?station=44034), is near the optimal identified location at meets all constraints (all bathymetry is on the edge). Dissolved oxygen level at this location is 9.8289mg/L, based on an interpolation of a data set from the National Centers for Environmental Information, but no time series data could be found. Data below is shown for the year 2022.
